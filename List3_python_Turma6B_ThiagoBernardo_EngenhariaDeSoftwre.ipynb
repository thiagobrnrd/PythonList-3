```python
#19 - Crie dois arrays de números inteiros aleatórios de tamanho 20 e calcule o ângulo entre eles (produto escalar normalizado).
array_1 = np.random.randint(1, 101, 20)
array_2 = np.random.randint(1, 101, 20)
produto_escalar = np.dot(array_1, array_2)
norma_1 = np.linalg.norm(array_1)
norma_2 = np.linalg.norm(array_2)
angulo_rad = np.arccos(produto_escalar / (norma_1 * norma_2))
angulo_deg = np.degrees(angulo_rad)

print(f"Ângulo entre os arrays (em graus): {angulo_deg}")
```

```python
#Exercícios de SciPy
#1 - Calcule a integral da fun¸c˜ao f(x) = x2 sin(x) no intervalo de 0 a π
import scipy.integrate as spi
import numpy as np

T = 10
integral, _ = spi.quad(lambda z: z**2 * np.sin(z), 0, T)
print("1. Integral:", integral)
```

```python
#2 - Resolva a equa¸c˜ao diferencial dy/dx = x−y, com y(0) = 1, no intervalo de 0 a 10.
import numpy as np

def diff_eq(y, x):
    return x - y
x_vals = np.linspace(0, 10, 100)
y0 = 1
y_vals = spi.odeint(diff_eq, y0, x_vals)
print("2. ED Solução:", y_vals[-1])
```

```python
#3 - Encontre as ra´ ızes da fun¸c˜ao f(x) = x4 − 3x3 + 2 utilizando SciPy
import scipy.optimize as spo
import numpy as np

def func(z):
    return z**4 - 3*z**3 + 2
raizes = spo.fsolve(func, [0, 1, 2, 3])
print("3. Raízes:", raizes)
```

```python
#4 - Use a fun¸c˜ao de otimiza¸ c˜ao de SciPy para maximizar f(x) = −x2+4x+1
import scipy.optimize as spo

max_result = spo.minimize(lambda z: -(-z**2 + 4*z + 1), 0)
print("4. Máximo:", -max_result.fun)
```

```python
#5 - Calcule a transformada de Fourier de um sinal composto de sin(2πt) e cos(4πt).
import scipy.fft as fft

x = np.linspace(0, 10, 100)
signal = np.sin(2 * np.pi * x) + np.cos(4 * np.pi * x)
fft_result = fft.fft(signal)
print("5. FFT:", fft_result[:5])
```

```python
#10 - Calcule a integral dupla da fun¸ c˜ao f(x,y) = xy + x2 no intervalo [0,2] para x e [0,1] para y.
import scipy.integrate as spi

def f(x, y):
    return x * y + x**2
x_min, x_max = 0, 2
y_min, y_max = 0, 1
resultado, erro = spi.dblquad(f, x_min, x_max, lambda x: y_min, lambda x: y_max)

print("Resultado da integral dupla:", resultado)
```

```python
#8 - Plote um gráfico de linha com intervalos de confiança (sombramento) para a função y = sin(x).

x = np.linspace(0, 10, 100)
y = np.sin(x)

erro = 0.2
y_sup = y + erro
y_inf = y - erro

plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r"$y = \sin(x)$", color="blue")
plt.fill_between(x, y_inf, y_sup, color="blue", alpha=0.2, label="Intervalo de Confiança")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Gráfico de Linha com Intervalo de Confiança")
plt.legend()
plt.grid(True)

plt.show()
```
